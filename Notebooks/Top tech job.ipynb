```python
fresher_jobs = df[df['experience_min_yrs']<=1]
```

```python
print(df['is_fresher_friendly'].head())
```

```python
fresher_df = df[df['is_fresher_friendly']==True]
```

```python
fresher_companies = fresher_df.groupby('company_name').size()
```

```python
fresher_companies = fresher_companies.sort_values(ascending= False)
```

```python
print(fresher_companies.head(20))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

fresher_companies.head(10).plot(kind='bar')

plt.title('Top 10 Fresher-Friendly Companies')
plt.ylabel('Number of Job Openings')
plt.xlabel('Company')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
domain_salary = (
    df.groupby('skill_domain')['salary_midpoint_lpa']
      .mean()
      .sort_values(ascending=False)
)

print(domain_salary)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

domain_salary.plot(kind='bar')

plt.title('Average Salary by Skill Domain')
plt.ylabel('Average Salary (LPA)')
plt.xlabel('Skill Domain')

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
city_jobs = (
    df.groupby('primary_city')
      .size()
      .sort_values(ascending=False)
      .head(15)
)

print(city_jobs)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

city_jobs.plot(kind='bar')

plt.title('Top 15 Cities by Hiring Demand')
plt.ylabel('Number of Job Openings')
plt.xlabel('City')

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
experience_salary = (
    df.groupby('experience_tier')['salary_midpoint_lpa']
      .mean()
      .sort_values()
)

print(experience_salary)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

experience_salary.plot(
    kind='line',
    marker='o'
)

plt.title('Experience vs Average Salary')
plt.ylabel('Average Salary (LPA)')
plt.xlabel('Experience Tier')

plt.grid(True)
plt.tight_layout()

plt.show()
```
